## Kluczowe pomysły w rozwiązaniu (wersja uproszczona)

### 1. „Samoleczenie” kontekstu
Problem: system ewaluacyjny ucina początek pliku, czyli często znika:
- sekcja importów,
- definicja głównej klasy.

To są krytyczne informacje, bez których model „nie wie”, w jakim środowisku działa.

Rozwiązanie:
- brakujący fragment z początku pliku jest ręcznie dodawany z powrotem do kontekstu,
- model dostaje więc pełniejszy obraz pliku, mimo ograniczeń.

---

### 2. Wyciąganie definicji (AST + Regex)
Trzeba wiedzieć, co każdy plik w repozytorium zawiera (funkcje, klasy).

Jak to działa:
- najpierw używany jest parser AST (czyli dokładna analiza składni Pythona),
- jeśli plik ma błąd i AST nie działa → system przełącza się na prostsze Regexy.

Efekt:
- mamy dokładność, gdy się da,
- i odporność na błędy, gdy kod jest niedokończony.

---

### 3. Lepsze wyszukiwanie niż „liczenie słów”
Zamiast prostego sprawdzania wspólnych słów:
- używany jest algorytm BM25 (jak w wyszukiwarkach).

Co to daje:
- rzadkie i ważne słowa mają większe znaczenie,
- popularne słowa (np. `data`, `value`) mają mniejszy wpływ,
- trafniej dobierany jest kontekst.

---

### 4. Mapa projektu
Do kontekstu dodawana jest uproszczona struktura folderów projektu.

Po co:
- model lepiej rozumie organizację kodu,
- łatwiej domyśla się, gdzie znajdują się różne elementy,
- poprawia to orientację w większych repozytoriach.

---

## Jak wybierane są najlepsze pliki

Każdy plik dostaje punkty według kilku zasad. Im więcej punktów, tym większa szansa, że trafi do kontekstu.

### Najważniejsze kryteria:

**1. Importy (najważniejsze)**
- jeśli plik jest bezpośrednio importowany → najwyższy priorytet,
- to praktycznie pewny wybór.

**2. Trafienie w definicję**
- jeśli plik zawiera funkcję lub klasę używaną przy kursorze → duży bonus,
- oznacza to, że model prawdopodobnie będzie jej potrzebował.

**3. Ogólne dopasowanie (BM25)**
- jeśli plik jest tematycznie podobny → dostaje dodatkowe punkty,
- pomaga dobrać kontekst uzupełniający.

**4. Bliskość w projekcie**
- pliki z tego samego folderu są trochę ważniejsze,
- bo zwykle są ze sobą powiązane.
